```python
# get datetime, title, place and message from csv file.
csv_name = 'shcedule.csv'
minutes = 3 # minutes before notifying.

with open(csv_name, encoding='utf-8', newline='') as f:
    text = csv.reader(f, delimiter=',')
    all_list = [row for row in text]

all_list1 = all_list[1:]
all_list2 = [] # [datetime, title, place, message]
for row in all_list1:
    dt_str = row[0][:-3] + ' ' + row[1][:5]
    dt_date = dt.datetime.strptime(dt_str, '%Y/%m/%d %H:%M') - delta_date
    del row[:2]
    if len(row)==0:
        row.extend([dt_date, '', '', ''])
        all_list2.append(row)
    elif len(row)==1:
        row.insert(0, dt_date)
        row.extend(['', ''])
        all_list2.append(row)
    elif len(row)==2:
        row.insert(0, dt_date)
        row.extend([''])
        all_list2.append(row)
    else:
        row.insert(0, dt_date)
        all_list2.append(row)

# プログラム実行時よりも古いスケジュールは削除する。
all_list3 = [] # [datetime_str, title, place, message]
now_dt = dt.datetime.now()
for row in all_list2:
    dt_date = row[0]
    dt_str = dt_date.strftime('%Y%m%d %H:%M')
    if dt_date > now_dt:
        row[0] = dt_str
        all_list3.append(row)

while all_list3 != []:
    now_str = dt.datetime.now().strftime('%Y%m%d %H:%M')
    for row in all_list3:
        dt_str, title, place, message = row
        if dt_str == now_str:
            notification.notify(
                title = title,
                message = f"時間: {dt_str[-5:]}  場所: {place}  メッセージ: {message}",
                app_name = "アプリ名",
                #app_icon="C:\office\office.ico",
                timeout=30
            )
            all_list3.remove(row)
    time.sleep(5)
```
